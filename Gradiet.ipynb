```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return 2*x**4 - 5*x**3- x**2 + 7*x + 18
def df(x):
    return 8*x**3-15*x**2-2*x+7
```

```python
x=np.arange(-5,5,0.01)

plt.plot(x,f(x))
plt.ylim(0,700)
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.plot(x,f(x))
k=5
plt.scatter(k,f(k))
constant=0.001
for i in range(100):
    k=k-constant*df(k)
    plt.scatter(k,f(k))
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rc('font', family='Noto Sans CJK TC')
```

```python
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/101/101_class.csv')
df.head()
```

```python
TPE = df[(df['縣市名稱'] == '30 臺北市') & (df['等級別']  == 'B 學士') & (df['日夜別']  == 'D 日')]
TPE.head()
```

```python
df = TPE.sort_values(by = ['四年級'], ascending=False)
df[:20].plot.bar(x='學校名稱', y=['四年級'])
```
